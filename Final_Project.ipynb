```python
def convertor(x):
    if(x > 1400):
        return 1
    else:
        return 0
```

```python
data['NumberOfShares'] = data[' shares'].apply(convertor)
```

```python
data.head()
```

```python
X = data.drop([' shares', 'NumberOfShares', 'url'], axis=1)
```

```python
y = data['NumberOfShares']
```

```python
# Spliting data into test, train set
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

# Logistic Regression - All predictors

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predict_y = logmodel.predict(X_test)
```

```python
print(classification_report(y_test,predict_y))
```

```python
print(confusion_matrix(y_test,predict_y))
```

```python
print(accuracy_score(y_test, predict_y))
```

# Generalized Linear Model

```python
glm_model = sm.GLM(y_train, X_train, family = sm.families.Binomial())
```

```python
results = glm_model.fit()
glm_y = results.predict(X_test)
```

```python
print(results.summary())
```

```python
from scipy import stats
```

```python
fig, ax = plt.subplots()
resid = results.resid_deviance.copy()
resid_std = stats.zscore(resid)
ax.hist(resid_std, bins = 5)
ax.set_title('Histogram of Standardized deviance residuals')
```
